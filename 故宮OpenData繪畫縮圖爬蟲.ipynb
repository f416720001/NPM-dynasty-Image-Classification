## 導入所需的爬蟲套件

```python
# !pip install bs4
# !pip install wget
```

```python
import requests
import pandas as pd
import os
from bs4 import BeautifulSoup
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import urllib.request
from urllib.parse import quote
 
def download(output, url):
    # utf-8 網址處理
    url = quote(url, safe='/:?=&')
    print(url)
    urllib.request.urlretrieve(url, filename=output)
    print(output)
```

## 爬蟲開始，找到所有縮圖->放到images->指定朝代資料夾->下載

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.135 Safari/537.36'}

for dynasty in links:
    # Create folder
    if not os.path.exists(dynasty["name"]):
        os.makedirs(dynasty["name"])

    print(dynasty["name"], dynasty["link"])
    url= dynasty["link"]
    
    count = 1
    for pageNumber in range(1, 400):
        
        print("pageNumber " + str(pageNumber))
        html = requests.get(url + "&pageNo=" + str(pageNumber), headers=headers)
        html.encoding = 'UTF-8'
        sp = BeautifulSoup(html.text, 'html.parser')
        
        if len(sp.find(class_ = "painting-list")) == 1:
            break

        for items in sp.find_all(class_ = "painting-list"):
            for item in items.find_all(class_ = "list"):
                print(item.h3.text, item.img["src"])
                path = "https://theme.npm.edu.tw/opendata/" + item.img["src"]
                try: 
                    download(dynasty["name"] + "/" + item.h3.text.strip() + "-" + str(count) + ".jpg", path)
                except:
                    print(item.h3.text, item.img["src"])
                count += 1
```
